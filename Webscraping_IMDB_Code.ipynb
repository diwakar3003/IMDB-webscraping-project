## Scraping

```python
# Note this takes about 40 min to run if np.arange is set to 9951 as the stopping point.

pages = np.arange(20, 161, 50) # Last time I tried, I could only go to 10000 items because after that the URI has no discernable pattern to combat webcrawlers; I just did 4 pages for demonstration purposes. You can increase this for your own projects.
headers = {'Accept-Language': 'en-US,en;q=0.8'} # If this is not specified, the default language is Chinese

#initialize empty lists to store the variables scraped
titles = []
years = []
ratings = []
genres = []
runtimes = []
imdb_ratings = []
imdb_ratings_standardized = []
metascores = []
votes = []

for page in pages:
     #print(str(page)+" PAGE ")
    #get request for sci-fi
     response = get("https://www.imdb.com/search/title/?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&" 
                   + "start=" 
                   + str(page) 
                   + "&ref_=adv_nxt", headers=headers)
     sleep(randint(8,15))
     
    #throw warning for status codes that are not 200
     if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))

    #parse the content of current iteration of request
     page_html = BeautifulSoup(response.text, 'html.parser')
        
     movie_containers = page_html.find('div', class_ = 'lister list detail sub-list').find('div',class_="lister-list").find_all('div',class_="lister-item mode-advanced")
    
    #extract the 50 movies for that page
     for container in movie_containers:

        #conditional for all with metascore
        if container.find('div', class_ = 'lister-item-content') is not None:
             
            if container.find('h3',class_="lister-item-header").find('span', class_= 'lister-item-index unbold text-primary') is not None:
              
              #year released
              rank = container.find('h3',class_="lister-item-header").find('span', class_= 'lister-item-index unbold text-primary').text.split('.')[0] # remove the parentheses around the year and make it an integer
              print(rank +" rank ")
              years.append(rank)
              if( int(rank)<50):
                 continue

            else:
              years.append(None) # each of the additional if clauses are to handle type None data, replacing it with an empty string so the arrays are of the same length at the end of the scraping

            if container.find('p',class_='text-muted ').find('span', class_ = 'runtime') is not None:
              
              #runtimes
              runtime = container.find('p',class_='text-muted ').find('span', class_ = 'runtime').text.split()[0]
              print(runtime+" RUN")
              runtimes.append(int(runtime))

            else:
              runtimes.append(None)   
            #title
            title = container.find('h3',class_="lister-item-header").a.text
            print(title+" TI")
            titles.append(title)
            


            

            



   
```
